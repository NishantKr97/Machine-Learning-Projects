```python
import numpy as np
from sklearn import preprocessing, cross_validation, neighbors
import pandas as pd
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_csv('cancer-wisconsin.data') #data downloaded from https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data
```

```python
df.fillna(value=-99999, inplace=True)               #Filling not number cells with large negative value
df.replace('?',-99999, inplace=True) #replacing unknown dataset with large negative value
```

```python
X = np.array(df.drop(['class'], 1)) #from dataset dropping column class which is the column to be predicted
```

```python
y = np.array(df['class']) # Creating the column to be predicted
```

```python
p = pd.DataFrame(X)
```

```python
p.plot()
```

```python
p = pd.DataFrame(y)
p.plot()
```

```python
print(X)
```

```python
print(y)
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.4) #Splitting dataset into training and test set
```

```python
p = pd.DataFrame((X_train))
p.plot()
```

```python
df.plot()
```

```python
clf = neighbors.KNeighborsClassifier() 
```

```python
clf.fit(X_train, y_train)   # Applying KNeighborsClassifier to training data
```

```python
accuracy = clf.score(X_test, y_test)    #score returns the accuracy of the trained data
```
